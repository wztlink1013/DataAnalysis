### np的函数：np.linspace(start, stop, num=50, endpoint=True, retstep=False, dtype=None, axis=0)
    Return evenly spaced numbers over a specified interval.
    (在start和stop之间返回均匀间隔的数据)
    Returns num evenly spaced samples, calculated over the interval [start, stop].
    (返回的是 [start, stop]之间的均匀分布)
    The endpoint of the interval can optionally be excluded.
    Changed in version 1.16.0: Non-scalar start and stop are now supported.
    (可以选择是否排除间隔的终点)

    start:返回样本数据开始点
    stop:返回样本数据结束点
    num:生成的样本数据量，默认为50
    endpoint：True则包含stop；False则不包含stop
    retstep：If True, return (samples, step), where step is the spacing between samples.(即如果为True则结果会给出数据间隔)
    dtype：输出数组类型
    axis：0(默认)或-1

```python
print(np.linspace(2.0, 3.0, num=5))
print(np.linspace(2.0, 3.0, num=5, endpoint=False))
print(np.linspace(2.0, 3.0, num=5, retstep=True))
```

## 1. 多项式拟合===一阶

```python
def f(x):
    return 2*x + 1
x = np.linspace(-5,5)#原始x值
y = f(x)+np.random.randn(len(x))#原始y值
y_fit=np.polyfit(x,y,1)#一次拟合
y_show=np.poly1d(y_fit)#展示出来拟合后的方程
print('多项式的系数（精确度较大）：')
print(y_fit)#展示的系数 ===由高项式系数往低项式依次列出
print('多项式方程：')
print(y_show)

def f_fit(x,y_fit):
    a,b=y_fit.tolist()
    return a*x+b
y1=f_fit(x,y_fit)#将拟合的参数值弄成y值


plt.plot(x,y,'.',c='g')
plt.plot(x,y1,'b')
plt.show()
```

## 2. 多项式拟合===二阶、四阶

```python
def f(x):
    return x**2+1
x=np.linspace(-5,5)
y=f(x)+np.random.randn(len(x))

y_fit=np.polyfit(x,y,2)#二次多项式拟合
y_show=np.poly1d(y_fit)#polyfit能展示出来多项式方程
print('多项式的系数（精确度较大）：')
print(y_fit)
print('多项式方程：')
print(y_show)

def f_fit(x,y_fit):
    a,b,c=y_fit.tolist()
    return a*x**2+b*x+c
y1=f_fit(x,y_fit)#将拟合的参数值弄成y值

plt.scatter(x,y,c='g',label='散点数据')#散点图
#plt.plot(x,f(x),'r',label='original')#最标准的===最最理想的拟合曲线
plt.plot(x,y1,'b',label='拟合曲线')#拟合曲线

plt.title('二次拟合')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```

```python
x = np.arange(1, 17, 1)
y = np.array([4.00, 6.40, 8.00, 8.80, 9.22, 9.50, 9.70, 9.86, 10.00, 10.20, 10.32, 10.42, 10.50, 10.55, 10.58, 10.60])

z1 = np.polyfit(x, y, 4)#用4次多项式拟合
p1 = np.poly1d(z1)
yvals=p1(x)#也可以使用yvals=np.polyval(z1,x)
print('y = \n',p1) #在屏幕上打印拟合多项式

plot1=plt.plot(x, y, '*',label='实际数据图像')
plot2=plt.plot(x, yvals, 'r',label='拟合图像')
plt.xlabel('x 轴')
plt.ylabel('y 轴')
plt.legend(loc='best')
plt.title('四阶拟合曲线')
plt.show()
```

## 2. 指定函数拟合===指数

```python
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
import numpy as np

x = np.arange(1, 17, 1)
y = np.array([4.00, 6.40, 8.00, 8.80, 9.22, 9.50, 9.70, 9.86, 10.00, 10.20, 10.32, 10.42, 10.50, 10.55, 10.58, 10.60])

#用指数形式来拟合
def func(x,a,b):
    return a*np.exp(b/x)
popt, pcov = curve_fit(func, x, y)
a=popt[0]#popt里面是拟合系数，读者可以自己help其用法
b=popt[1]
yvals=func(x,a,b)
print('拟合最优系数分别为：\n',popt)#最优参数
print('拟合最优协方差估计矩阵分别为：\n',pcov)#最优参数的协方差估计矩阵

plot1=plt.plot(x, y, '*',label='实际数据图像')
plot2=plt.plot(x, yvals, 'r',label='拟合图像')
plt.xlabel('x 轴')
plt.ylabel('y 轴')
plt.legend(loc='best')
plt.title('指定函数（指数）拟合曲线')
plt.show()
```

```python
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
import numpy as np

x=np.linspace(-5,5)
y=x*x+1+np.random.randn(len(x))

#建立起一个指数函数模型
def func(x,a,b,c):
    return a*x*x+b*x+c

popt, pcov = curve_fit(func, x, y)#参数1：放入函数模型，参数2&3：放入要拟合的x，y数据，返回的是拟合的最后系数等
a=popt[0]#popt里面是拟合系数
b=popt[1]
c=popt[2]
yvals=func(x,a,b,c)

print('拟合最优系数分别为：\n',popt)#最优参数
print('拟合最优协方差估计矩阵分别为：\n',pcov)#最优参数的协方差估计矩阵

plot1=plt.plot(x, y, '*',label='实际数据图像')
plot2=plt.plot(x, yvals, 'r',label='拟合图像')
plt.xlabel('x 轴')
plt.ylabel('y 轴')
plt.legend(loc='best')
plt.title('指定函数(二次)拟合曲线')
plt.show()
```
