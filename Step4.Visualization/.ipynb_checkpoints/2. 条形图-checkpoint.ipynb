# 条形图
    1. 条形图和水平条形图
    2. 双条形图
    3. 堆叠条形图
    4. 其他种类条形图

```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib
%matplotlib inline
matplotlib.rcParams['font.sans-serif']=['SimHei']   
matplotlib.rcParams['axes.unicode_minus']=False     
```

### 1. 条形图bar和水平条形图barh

```python
x = [1,2,3,4,5]
y = [2,5,1,5,3]
power = [100,98,95,95,90]
variance = [0.1,0.3,0.2,0.1,0.1]
fig,ax = plt.subplots(ncols = 2)

ax[0].bar(x,y,color = 'cyan',yerr=variance)#align='center'啥意思？

ax[1].barh(range(5), power, height=0.7, color='blue', alpha=0.7)      # 下面是坐标原点；height是条形的宽度
ax[1].set_xlim(50,110)
#plt.yticks(range(5), ['逆鳞', '残虹', '渊虹', '鲨齿', '非攻'])
ax[1].set_yticks(range(5))  
ax[1].set_yticklabels(['逆鳞', '残虹', '渊虹', '鲨齿', '非攻'],rotation=30)

plt.show()
```

### 2. 双条形图

```python
label_list=['韩非||逆鳞','张良','卫庄','嬴政']
num_list1=[100,70,98,90]
num_list2=[99,95,90,90]
wucha = [1,5,10,10]
fig,ax = plt.subplots(figsize=(10,8))
'''基本操作'''
width = 0.3
x_1 = list(range(len(num_list1)))
x_2 = [i + width for i in x_1]
rects1=plt.bar(x_1,num_list1,width,alpha=0.5,color ='blue',label='武力值')
rects2=plt.bar(x_2,num_list2,width,alpha=0.5,color='#00ffff',yerr=wucha,label='智力值')
plt.legend()

'''条形上注释'''
for rect in rects1:
    height=rect.get_height()
    plt.text(rect.get_x() + rect.get_width() / 2, height+1, str(height),fontsize = 18, ha="center", va="bottom")
for rect in rects2:
    height = rect.get_height()
    plt.text(rect.get_x() + rect.get_width() / 2, height+1, str(height), fontsize = 18,ha="center", va="bottom")
'''基准线'''
plt.hlines(min(num_list1),-1,len(num_list1)+0.5,linestyle='dashed')#如果是水平条形图===hline改成vlines
                                                                    #plt.ylim(-1,len(num_list1)+0.1)
plt.axhline(100,color='red',linewidth=1)
plt.axvline(1,color='yellow',linewidth=2)
'''分区域'''
for bar,height in zip(rects1,num_list1):
    if height<91:
        bar.set(edgecolor = 'none',color = 'red',linewidth = 3)

'''x轴标签'''
plt.xticks([index+0.15 for index in x_1],label_list)   

plt.show()
```

### 3. 堆叠条形图
    ps：常见的bug===sitional argument follows keyword argument位置参数跟随关键字参数，说明了不要再bar里面加上height=num_list1

```python
label_list=['韩非||逆鳞','张良','卫庄','嬴政']
num_list1=[100,70,60,50]
num_list2=[84,92,95,98]

width=0.45
x = np.arange(len(num_list1))
rects3 = plt.bar(x, num_list1, width, alpha=0.8, color='red')
rects4 = plt.bar(x, num_list2, width, color='green', bottom=num_list1)
plt.legend()

for rect1 in rects4:
    for rect in rects3:
        height1=rect.get_height()
        height2=rect1.get_height()
        height = height1+height2
        plt.text(rect1.get_x() + rect1.get_width() / 2, height+1, str(height),fontsize = 18, ha="center", va="bottom")
        break
    continue
    
plt.ylim(0,220)
plt.xticks(x, label_list)
plt.xlabel("人物")
plt.ylabel("技力值")
plt.title("天行九歌")
plt.show()
```

### 4. 其他种类条形图

```python
mean_values = [1,2,3]
variance = [0.2,0.4,0.5]
bar_label = ['bar1','bar2','bar3']

x_pos = list(range(len(bar_label)))
plt.bar(x_pos,mean_values,yerr=variance,alpha=0.3)
max_y = max(zip(mean_values,variance))

plt.ylim([0,max_y[0]+max_y[1]*1.2])
plt.ylabel("variance y")
plt.xticks(x_pos,bar_label)
plt.show()
```
