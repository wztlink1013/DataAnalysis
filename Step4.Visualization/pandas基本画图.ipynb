# （四）、Pandas的绘图函数
#### 1.折线图
#### 2.水平/垂直/堆叠条形图
#### 3.散点矩阵

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.pyplot 
%matplotlib inline
matplotlib.rcParams['font.sans-serif']=['SimHei']   # 用黑体显示中文=====不要改matplotlib里面的配置
matplotlib.rcParams['axes.unicode_minus']=False     # 正常显示负号
```

## 1. 折线图

### ①Series
    plot的函数：label\ax\style\alpha\kind\logy\use_index\rot\xticks\xlim\grid

```python
ser_obj = pd.Series([2,4,2,5,7,12,54],index = ['一','二','三','四','五','六','七'])

ser_obj.plot()
```

### ②DataFrame

```python
array = [[1,2,3,4,5],
         [5,4,3,2,1],
         [7,3,5,2,8]]
df_obj = pd.DataFrame(array, columns = ['a', 'b', 'c', 'd', 'e'],index = ['one','two','three'])

df_obj.plot()
```

```python
#假如有一个DataFrame类型的数据 === 想要以一列为横轴，另一列为纵轴画图的话。
fig,ax = plt.subplots(ncols = 2)
ax[0].plot(df_obj['a'], df_obj['b'])
```

## 2. 水平/垂直/堆叠条形图

```python
print(ser_obj)
print(df_obj)
```

```python
ser_obj.plot(kind='bar')
df_obj.plot(kind='bar')
df_obj.plot(kind='barh')
df_obj.plot(kind='barh',stacked=True, alpha=1)
```

## 3. 散点矩阵

```python
print(df_obj)
#方式一
df_obj.plot.scatter(x='b',y='c')

#方式二
df_obj.plot(kind="scatter",x="a",y="b")
```
